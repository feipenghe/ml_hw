```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 64
Z = np.random.rand(n)
Z = [z*np.random.choice([-1, 1])*3 for z in Z] # [-3, .., 0, .., 3]
print(Z)
plt.xlim(-3, 3)
plt.step(sorted(Z), np.arange(1, n+1)/float(n))
```

```python
x = -5
Fn_x_hat = sum([z <= x for z in Z])
print(Fn_x_hat)
```

```python
import math
import matplotlib.pyplot as plt
import numpy as np
ks = [1,8,64,512]
n = round(1/(0.0025**2 *4))
for k in ks:
    ns = np.sum(np.sign(np.random.randn(n, k))*np.sqrt(1./k), axis=1)
    plt.plot(np.sort(ns), np.arange(1, n+1)/float(n), label = str(k))
plt.step(sorted(np.random.randn(n)), np.arange(1, n+1)/float(n)) 
plt.legend(loc = 1)
plt.xlabel("Observations")
plt.ylabel("Probability")
print(n)
```

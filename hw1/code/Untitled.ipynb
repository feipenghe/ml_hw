```python
import numpy as np
import math
from matplotlib import pyplot as plt
n = 256
a = np.linspace(0, 1,n)  # determinastically sample data from 0-1

ms = np.array([1,2,4,8,16,32])


sigma = math.sqrt(1)
f = 4*np.sin(math.pi*a)*np.cos(6*math.pi*a**2)  # true function
epsilon = np.random.normal(0, sigma, size=(256))  # noise
ys = f +epsilon

emp_errs = []
biass = []
variances = []
for m in ms:
    step_f = np.repeat( np.mean(ys.reshape(-1, m), axis = 1), m)  # function approximator for different m
    emp_err = (1/n)*np.sum((step_f-f)**2)
    emp_errs.append(emp_err) 
    bias = 0
    variance = 0
    # each represents on iteration over data
    for j in range(int(n/m)):  # iterate over batches
        
        i_0 = j*m
        i_end = (j+1)*m
        
        # bias
        f_avg_j = np.average(f[i_0:i_end])    # true model output average in a range
        f_j = f[i_0:i_end]  # true data points (y)
        bias += np.sum( (f_avg_j - f_j)**2 )
        
        # variance
        c_j = np.average(ys[i_0:i_end]) # average y_hat
        variance +=  m* np.average((  c_j - f_avg_j   )**2)

    avg_bias = (1/n)*bias
    variance = (1/n)*variance
    biass.append(avg_bias)
    variances.append(variance)


plt.plot(ms,emp_errs, label = "empirical error")
plt.plot(ms, biass, label = "average bias-squared")
plt.plot(ms, variances, label = "average variance")
plt.plot(ms, np.add(biass,variances), label = "error sum")
plt.xlabel("m")
plt.ylabel("errors")
plt.legend(loc = 0)
plt.show()
```

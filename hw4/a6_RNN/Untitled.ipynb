```python
import os


sst_path = "./data/SST-2"
sst_name = "train.tsv"


sst_f_pos = os.path.join(sst_path, sst_name)
```

```python
import csv
with open(sst_f_pos) as f:
    reader = csv.reader(f, delimiter = "\t")
    sents = []
    labels = []
    for row in reader:
        if row[0] == 'sentence':
            continue # skip titles
        sents.append(row[0])
        sents.append(row[1])


```

```python
max_sequence_length = -1
for sent in sents:
    temp_len = len(sent.split())
    if temp_len > max_sequence_length:
        max_sequence_length = temp_len
print(max_sequence_length)
```

```python
pad_token = ['<pad>']

a = ["a"]
padded_sents = []
for i in range(len(sents)):
    cur_sent_l = sents[i].split()
    num_pad = max_sequence_length - len(cur_sent_l)
    padded_sents.append(cur_sent_l + num_pad * pad_token )
```

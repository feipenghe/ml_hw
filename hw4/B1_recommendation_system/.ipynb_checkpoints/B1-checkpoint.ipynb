```python

import csv
import numpy as np
data = []
with open('./ml-100k/u.data') as csvfile:
    spamreader = csv.reader(csvfile, delimiter='\t')
    for row in spamreader:
        data.append([int(row[0])-1, int(row[1])-1, int(row[2])])
data = np.array(data)
num_observations = len(data)
num_users = max(data[:,0])+1
num_items = max(data[:,1])+1
# num_observations = 100,000
# num_users = 943, indexed 0,...,942
# num_items = 1682 indexed 0,...,1681
np.random.seed(1)
num_train = int(0.8*num_observations)
perm = np.random.permutation(data.shape[0])
train = data[perm[0:num_train],:]
test = data[perm[num_train::],:]
```

```python
import pandas as pd
train

train_df =  pd.DataFrame(train, columns = ["user_idx", "movie_idx", "user_score"] )
test_df = pd.DataFrame(test, columns = ["user_idx", "movie_idx", "user_score"])
```

```python
train_mean_score_df = train_df[["movie_idx", "user_score"]].groupby("movie_idx").mean()
test_mean_score_df = test_df[["movie_idx", "user_score"]].groupby("movie_idx").mean()

mean_score_df = train_mean_score_df.merge(test_mean_score_df, on = "movie_idx", how = "right", suffixes=["_train", "_test"]).fillna(0)

se =  (mean_score_df['user_score_train'] - mean_score_df['user_score_test']).to_numpy()**2
mse = (1/len(test)) * np.sum(se)
```

```python
def data2nparr(data):
    R = np.zeros((num_items, num_users))

    for d in data:
        user_idx = d[0]
        movie_idx = d[1]
        rating = d[2]
        R[movie_idx, user_idx] = rating
    return R

R = data2nparr(train)
```

```python
from scipy.sparse.linalg import svds, eigs
ds = [1,2,5,10,20,50]
log_test_mse = []
log_train_mse = []
for d in ds:
    print("d = ", d)
    # u, s, vt = svds(R)
    u, s, vt = np.linalg.svd(R)
    u = u[:, :d]
    s = np.diag(s[:d])
    vt = vt[:d,:]
    recon_R = u@ s @ vt

    test_mse = 0
    train_mse = 0

    # train
    for d in train:
        user_idx = d[0]
        movie_idx = d[1]
        rating = d[2]
        pred = recon_R[movie_idx][user_idx]
        train_mse += (pred - rating)**2

    # test
    for d in test:
        user_idx = d[0]
        movie_idx = d[1]
        rating = d[2]
        pred = recon_R[movie_idx][user_idx]
        test_mse += (pred - rating)**2

    test_mse = (1/len(test)) * test_mse
    train_mse = (1/len(train))* train_mse
    log_test_mse.append(test_mse)
    log_train_mse.append(train_mse)

from matplotlib import pyplot as plt

plt.plot(ds, log_test_mse, label = "test mse")
plt.plot(ds, log_train_mse, label = "train mse")
plt.legend()
plt.savefig("B1b")
```

```python
# not R as it initialize Nan with 0
step_size = 1e-5
reg_lambda = 1


log_train_am_loss = []
log_test_am_loss = []
for d in ds:
    print(d)
    U = np.random.rand(num_items, d) # movie
    V = np.random.rand(num_users, d)

    for data in train:
        j = data[0] # user
        i = data[1] # movie
        rating = data[2]
        while np.max(U[i].dot(V[j])-rating) > 0.5:
            # update u_i
            u_gradient = 2 * ( U[i].dot(V[j]) - rating) * V[j]  + 2 * reg_lambda * U[i]
            U[i]  -=  u_gradient * step_size

            # update V_j
            v_gradient = 2 * ( U[i].dot(V[j]) - rating) * U[i] + 2*reg_lambda* V[j]
            V[j] -=  v_gradient * step_size

    train_loss = 0
    test_loss = 0
    # training loss
    for data in train:
        j = data[0]
        i = data[1]
        rating = data[2]
        train_loss += (U[i].dot(V[j]) - rating)**2
    train_loss = train_loss/len(train)
    log_train_am_loss.append(train_loss)

    # test
    for data in test:
        j = data[0]
        i = data[1]
        rating = data[2]

        pred = U[i].dot(V[j])
        test_loss += (U[i].dot(V[j]) - rating)**2
    test_loss = test_loss/len(test)
    log_test_am_loss.append(test_loss)

    print("train loss: ", train_loss)
    print("test loss: ", test_loss)
```

```python
plt.plot(ds, log_train_am_loss, label = "train mse")
plt.plot(ds, log_test_am_loss, label = "test mse")
plt.legend()
plt.savefig("B1c")
```

```python
import torch
from torch import nn
class RegressionModel(nn.Module):
    def __init__(self, num_items, num_users):
        super().__init__()
        embedding_dim = 10
        # vocab_size = embedding_matrix.shape[0]
        # embedding_dim = embedding_matrix.shape[1]
        # Construct embedding layer and initialize with given embedding matrix. Do not modify this code.
        # self.N = 50  # N is number of sequences which are batch size
        self.embd1 = nn.Embedding(num_items,embedding_dim=10)
        self.embd2 = nn.Embedding(num_users, embedding_dim = 10)
        self.fc1 = nn.Linear(10+10, 100)
        self.fc2 = nn.Linear(100, 1)
        self.criterion = nn.MSELoss()
        self.relu = nn.Softmax()

    def forward(self, inputs):
        """
        Takes in a batch of data of shape (N, max_sequence_length). Returns a tensor of shape (N, 1), where each
        element corresponds to the prediction for the corresponding sequence.
        :param inputs: Tensor of shape (N, max_sequence_length) containing N sequences to make predictions for.
        :return: Tensor of predictions for each sequence of shape (N, 1).
        """
        input0 = self.embd1(inputs[0])
        input1 = self.embd2(inputs[1])
        inputs = torch.cat((input0, input1))
        # inputs -> embedding
        out = self.fc1(inputs)
        out = self.fc2(out)
        out = self.relu(out)
        return out

    def loss(self, inputs, targets):
        """
        Computes the binary cross-entropy loss.
        :param logits: Raw predictions from the model of shape (N, 1)
        :param targets: True labels of shape (N, 1)
        :return: Binary cross entropy loss between logits and targets as a scalar tensor.
        """
        return self.criterion(inputs, targets.float())

    # def accuracy(self, ):

model = RegressionModel(num_items, num_users)
optimizer = torch.optim.Adam(model.parameters(), lr = 0.05, weight_decay = 0.05)

count = 0
for data in train:
    j = data[0]
    i = data[1]

    optimizer.zero_grad()
    rating = torch.tensor(data[2])
    t = torch.tensor([i,j]).long()
    out = model.forward(t)
    loss = model.loss(out, rating)
    loss.backward()
    optimizer.step()
    if count % 1000 == 0:
        num_correct = 0
        with torch.no_grad():
            for data in test:
                j = data[0]
                i = data[1]
                rating = data[2]
                t = torch.tensor([i,j]).long()
                out = model.forward(t)
                print("out: ", out)
                print("rating: ", rating)
                if out == rating:
                    num_correct += 1
        print("test accuracy: ", num_correct/len(test))



    count += 1
```
